```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

salaries=pd.read_csv('datascience_salaries.csv')
salaries.head()

```

```python
salaries_copy=salaries.copy()
salaries_copy.head()
scaler = MinMaxScaler()

salaries_copy[['salary']] = scaler.fit_transform(salaries_copy[['salary']])
salaries_copy.head()
print(salaries_copy.dtypes)
print(salaries_copy.select_dtypes(include=['object']).columns.tolist())

```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

categorical_cols = salaries_copy.select_dtypes(include=['object']).columns

for col in categorical_cols:
    le = LabelEncoder()
    salaries_copy[col] = le.fit_transform(salaries_copy[col].astype(str))

salaries_copy[['job_title']] = scaler.fit_transform(salaries_copy[['job_title']])
salaries_copy[['job_type']] = scaler.fit_transform(salaries_copy[['job_type']])
salaries_copy[['experience_level']] = scaler.fit_transform(salaries_copy[['experience_level']])
salaries_copy[['location']] = scaler.fit_transform(salaries_copy[['location']])
salaries_copy[['salary_currency']] = scaler.fit_transform(salaries_copy[['salary_currency']])
salaries_copy.head()

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(salaries_copy)
print(reduced_data)
```
